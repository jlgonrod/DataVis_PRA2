# 0. Librerías

```python
import os
```

Se establece la ruta la carpeta y se comprueba que esté vacía

```python
# Establecemos el path de la carpeta de datos
data_folder = os.path.join("..", "0_data")

# Si hay algun archivo o carpeta 
if os.path.exists(data_folder):
    os.system("rm -r " + data_folder)
    os.mkdir(data_folder)

# Si no existe la carpeta la creamos
else:
    os.mkdir(data_folder)
```

# 1. Datos de MMASH

Descargamos el zip con los datos de https://physionet.org/content/mmash/1.0.0/

```python
# Descargamos los datos
os.system("wget -r -N -c -np https://physionet.org/files/mmash/1.0.0/ -P " + data_folder)
```

```python
# Movermos el archivo zip a la carpeta data
os.system("mv " + os.path.join(data_folder, "physionet.org", "files", "mmash", "1.0.0", "MMASH.zip") + " " + data_folder)

# Eliminamos el resto de archivos
os.system("rm -r " + os.path.join(data_folder, "physionet.org"))
```

```python
# Definimos el root y la estructura de ficheros
root = data_folder
estructura_de_ficheros = os.listdir(root)

# Imprimimos la estructura de ficheros
print(f"Root: {root}")
print("Estructura de ficheros:")
for fichero in estructura_de_ficheros:
    print(f"- {fichero}")
```

```python
# Descomprimimos el archivo usando una libreria de python
import zipfile

with zipfile.ZipFile(os.path.join(data_folder, "MMASH.zip"), 'r') as zip_ref:
    zip_ref.extractall(data_folder)

# Eliminamos el archivo zip
os.system("rm " + os.path.join(data_folder, "MMASH.zip"))

# Cambiamos el nombre de la carpeta DataPaper a MMASH
os.system("mv " + os.path.join(data_folder, "DataPaper") + " " + os.path.join(data_folder, "MMASH"))
```
